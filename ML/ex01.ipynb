# 선형 회귀

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./data/LinearRegressionData.csv')
df.head()
```

```python
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
X, y
```

```python
from sklearn.linear_model import LinearRegression as lr
reg=lr()
reg.fit(X, y)
```

```python
y_pred=reg.predict(X)
y_pred
```

```python
y
```

```python
X
```

### 선형회귀 시각화

```python
plt.figure(figsize=(10,5), dpi=75)
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='green')
plt.title('Score by hours')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
reg.predict([[9],[11], [8]])
```

```python
m=reg.coef_
```

```python
b=reg.intercept_
b
```

```python
y=m*9+b
y
```

### 데이터 세트 분리

```python
df=pd.read_csv('./data/LinearRegressionData.csv')
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split as ttp
X_train, X_test, y_train, y_test=ttp(X, y, test_size=0.2, random_state=0)
```

```python
X_train, len(X_train)
```

```python
X, len(X)
```

```python
y_train, len(y_train)
```

```python
X_test, len(X_test), y_test, len(y_test)
```

```python
from sklearn.linear_model import LinearRegression as lr
reg=lr()
reg.fit(X_train, y_train)
```

```python
y_pred=reg.predict(X_train)
plt.figure(figsize=(10,5), dpi=75)
plt.scatter(X_train, y_train, color='slateblue')
plt.plot(X_train, y_pred, color='orange')
plt.title('Score by hours (train data)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
plt.figure(figsize=(10,5), dpi=75)
plt.scatter(X_test, y_test, color='slateblue')
plt.plot(X_train, y_pred, color='orange')
plt.title('Score by hours (test data)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.score(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

```python
from sklearn.linear_model import SGDRegressor
sr=SGDRegressor(max_iter=1000, eta0=0.0001, random_state=0, verbose=1)
sr.fit(X_train, y_train)
```

```python
y_pred=sr.predict(X_train)
plt.figure(figsize=(10,5), dpi=75)
plt.scatter(X_train, y_train, color='slateblue')
plt.plot(X_train, y_pred, color='orange')
plt.title('Score by hours (SDG data)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
sr.score(X_train, y_train)
```
