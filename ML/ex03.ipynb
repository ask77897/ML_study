### 다항회귀

```python
import pandas as pd
```

```python
df=pd.read_csv('./data/PolynomialRegressionData.csv')
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
X, y
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X, y)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5), dpi=75)
plt.scatter(X, y, color='slateblue')
plt.plot(X, reg.predict(X), color='orange')
plt.title('Score by hours')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()

```

```python
reg.score(X, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=2)
X_poly=poly_reg.fit_transform(X)
```

```python
X_poly[:5], X[:5]
```

```python
reg=LinearRegression()
reg.fit(X_poly, y)
```

```python
reg.score(X_poly, y)
```

```python
#데이터 읽기
df=pd.read_csv('./data/PolynomialRegressionData.csv')
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
#X독립변수 다항으로 변경
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=2)
X_poly=poly_reg.fit_transform(X)
#선형객체만들어서 학습
reg=LinearRegression()
reg.fit(X_poly, y)
#X범위 0.1 단위로 생성
import numpy as np
X_range=np.arange(min(X), max(X), 0.1)
X_range=X_range.reshape(-1, 1)
#그래프 출력
plt.figure(figsize=(10,5), dpi=75)
plt.scatter(X, y, color='slateblue')
plt.plot(X_range, reg.predict(poly_reg.fit_transform(X_range)), color='orange')
plt.title('Score by hours')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
pred = reg.predict(poly_reg.fit_transform([[4.9187669647]]))
pred
```
