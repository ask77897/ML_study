```python
import pandas as pd
```

```python
df=pd.read_csv('./data/QuizData.csv')
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
len(X_train), len(X_test)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_train, y_train, color='mediumpurple')
plt.plot(X_train, reg.predict(X_train), color='orange')
plt.show()
```

```python
plt.scatter(X_test, y_test, color='mediumpurple')
plt.plot(X_train, reg.predict(X_train), color='orange')
plt.show()
```

```python
reg.score(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

```python
X_train, y_train
```

```python
from sklearn.preprocessing import PolynomialFeatures
reg_poly=PolynomialFeatures(degree=2)
X_poly=reg_poly.fit_transform(X_train)
X_poly
```

```python
reg=LinearRegression()
reg.fit(X_poly, y_train)
```

```python
import numpy as np
X_arange=np.arange(min(X_train), max(X_train), 0.1)
```

```python
plt.scatter(X_train, y_train, color='mediumpurple')
plt.plot(X_arange, reg.predict(X_poly), color='orange')
plt.show()
```

```python
df=pd.read_csv('./data/QuizData.csv')
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
```

```python
import numpy as np
X_arange=np.arange(min(X), max(X), 0.1)
X_arange= X_arange.reshape(-1, 1)
```

```python
X_poly = poly.fit_transform(X)
X_poly
```

```python
reg=LinearRegression()
reg.fit(X_poly, y)
```

```python
plt.figure(figsize=(10,5), dpi=70)
plt.scatter(X, y, color='green')
plt.plot(X_arange, reg.predict(poly.fit_transform(X_arange)), color='orange')
plt.show()
```

```python
reg.score(X_poly, y)
```

```python
reg.predict(poly.fit_transform([[300]]))
```
